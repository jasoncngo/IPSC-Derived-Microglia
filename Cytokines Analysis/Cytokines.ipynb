```python
import xlrd
import numpy as np
import pandas as pd
from scipy import stats
import csv
import matplotlib
from matplotlib import pyplot as plt
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
```

```python
textfiles=['./Oct-30-2021-Human-Cytokine .xlsx', 
           './Dec-2-2021-Human-Cytokine-71-Plex-Sher-37007-Tissue-cell-Culture-Supernatant (1).xlsx',
           './Aug-19-2022-Human-Cytokine.xlsx']
```

```python
for t in range(len(textfiles)):
    final=[]
    sheet = xlrd.open_workbook(textfiles[t]).sheet_by_index(0)
    for r in range(len(sheet.row_values(30))):
        if 'Units: pg/mL' in sheet.col_values(r):
            name=[(sheet.col_values(r)[(sheet.col_values(r).index('Units: pg/mL'))+1])]
            templist=(sheet.col_values(r-1)[((sheet.col_values(r).index('Units: pg/mL'))+3)+8:])
            templist2=[]
            dnolps=[]
            for k in range(len(templist)):
                if type(templist[k])==float:
                    templist2.append(templist[k])
                    if k>2:
                        dnolps.append(templist[k])
            if templist2!=[]:
                if t==0:
                    if type(templist[0])==float:
                        templist2=np.array(templist2)/templist[0]
                if t==1:
                    if type(templist[6])==float:
                        templist2=np.array(templist2)/templist[6]
                if t==2:
                    templist2=np.array(templist2)/np.mean(dnolps)
            final.append(name+list(templist2))
    if t==0:
        df = pd.DataFrame(final, columns=[0]+[str(t+1)+'-'+str(k+1)for k in range(len(final[0])-1)]).set_index(0)
    if t>0:
        df = df.join(pd.DataFrame(final, columns=[0]+[str(t+1)+'-'+str(k+1)for k in range(len(final[0])-1)]).set_index(0))
df
```

```python
nolpsdf=df[["1-1", "2-7","3-4","3-5","3-6"]].T
nolpsdf
```

```python
lpsdf=df[["1-2", "2-4","3-1","3-2","3-3"]].T
lpsdf
```

```python
pval=[]
l2fc=[]
lpsind=[]
nolpsind=[]
for k in range(len(list(nolpsdf.columns))):
    fnolps=[]
    flps=[]
    pfnolps=list(nolpsdf[list(nolpsdf.columns)[k]])
    pflps=list(lpsdf[list(nolpsdf.columns)[k]])
    for m in range(len(pfnolps)):
        if pfnolps[m]>0:
            fnolps.append(pfnolps[m])
        if pflps[m]>0:
            flps.append(pflps[m])
    pval.append(-np.log10(stats.ttest_1samp(flps,1)[1]))
    l2fc.append(np.log2(np.mean(flps)/np.mean(fnolps)))
    lpsind.append(np.mean(flps))
    nolpsind.append(np.mean(fnolps))
```

```python
fig, ax = plt.subplots()
ax.scatter(l2fc,pval, marker='.', s=10,c='b')
#plt.axhline(y=-np.log10(0.05), color='black', linestyle='dashed', linewidth=1)
plt.axvline(x=0, color='black', linestyle='dashed', linewidth=1)
for i,txt in enumerate(list(nolpsdf.columns)):
    if abs(l2fc[i])>=3:
        ax.annotate(txt, (l2fc[i]-0.4,pval[i]+0.1), size=10)
    if abs(pval[i])>=2.2:
        ax.annotate(txt, (l2fc[i]+0.04,pval[i]-0.1), size=10)
    if abs(l2fc[i])>=0.47 and abs(l2fc[i])<=3.5:
        if txt=='IL-3':
            ax.annotate(txt, (l2fc[i],pval[i]-0.3), size=10)
        elif txt=='TNFα':
            ax.annotate(txt, (l2fc[i],pval[i]-0.06), size=10)
        elif txt=='G-CSF':
            ax.annotate(txt, (l2fc[i],pval[i]+0.06), size=10)
        else:
            ax.annotate(txt, (l2fc[i],pval[i]+0.04), size=10)
ax.set_yscale('log')
ax.set(ylim=(min(pval),3.3), xlim=(-0.25,4.25))
plt.xlabel('log2fc')
plt.ylabel('-log10(p-value)')
plt.title(stats.ttest_rel(lpsind, nolpsind))
plt.savefig('LPSvsnoLPScombineddata.pdf')
```

```python
final=[['Cytokine', 'FI-LPS/FInoLPS','-log10pvalue', 'log2fc']]
for k in range(len(list(nolpsdf.columns)[1:])):
    final.append([list(nolpsdf.columns)[1:][k], list(lpsdf[list(nolpsdf.columns)[k+1]]), pval[k], l2fc[k]])
with open('LPSvsnoLPScombineddata.csv', 'w', newline='') as file:
    for k in range(len(final)):
        csv.writer(file).writerow(final[k])
final
```
